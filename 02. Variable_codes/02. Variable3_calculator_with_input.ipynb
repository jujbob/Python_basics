# Variable 기초 예제 #
 - 주제: input 함수를 활용한 사측연산 계산기 만들기
 - 방법: 사용자로부터 숫자 두개를 입력으로 받아 더하기, 빼기, 곱하기, 나누기 결과를 출력한다

```python
num1 = int(input("첫번째 숫자를 입력하세요==> "))
```

```python
num2 = int(input("두번째 숫자를 입력하세요==> "))
```

```python
num1
```

```python
result1 = num1 + num2
result2 = num1 - num2
result3 = num1 * num2
result4 = num1 / num2
result5 = num1 % num2
result6 = num1 ** num2
```

```python
print(num1, "+", num2, "=", result1)
print(num1, "-", num2, "=", result2)
print(num1, "*", num2, "=", result3)
print(num1, "/", num2, "=", result4)
print(num1, "%", num2, "=", result5)
print(num1, "**", num2, "=", result6)
```
