# Variable 기초 예제 #
 - 주제: 사측연산 계산기 만들기
 - 방법: 변수 두개를 입력으로 받아 더하기, 빼기, 곱하기, 나누기 결과를 출력한다

print("hello, pytho")

```python
num1 = 100
num2 = 200
```

```python
print(num1)
print(num2)
```

```python
result1 = num1 + num2
```

```python
result2 = num1 - num2
```

```python
result3 = num1 * num2
```

```python
result4 = num1 / num2
```

```python
print(num1, "+", num2, "=", result1)
print(num1, "-", num2, "=", result2)
print(num1, "*", num2, "=", result3)
print(num1, "/", num2, "=", result4)
```
