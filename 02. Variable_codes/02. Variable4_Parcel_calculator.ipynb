# Variable 기초 예제 #
 - 주제: input 함수를 활용한 택배 요금 계산기 만들기
 - 방법: 사용자로부터 이름, 주소, 물건의 무게를 입력으로 받아 택배 요금을 반환하는 프로그램을 작성하시오.

```python
userName = input("보내는 사람 이름: ")
```

```python
userAdd = input("받을 사람의 주소를 입력하세요: ")
```

```python
parcel_weight = input("물건의 무게를 입력 하시오: ")
```

```python
par_price = 2000
```

```python
print("받는사람:", userName, "받는사람 주소:", userAdd, "택배 비용: ", int(parcel_weight)*par_price)
```
