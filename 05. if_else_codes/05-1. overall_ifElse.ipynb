# 05. 조건문 실습
 - 조건문의 3가지 다른 형태를 실습한다.
  *   if
  *   if - else
  *   if - elif - else



```python
num = input("숫자를 입력하세요")
```

```python
num
```

```python
if int(num) > 100:
  print("100보다 큽니다")
else:
  print("100보다 작습니다")

print("int(num > 100)의 값이", str(int(num) > 100), "입니다.")
print("코드 종료")
```

## 홀수, 짝수를 구분하는 프로그램

```python
if int(num) % 2 == 0:
  print("짝수 입니다.")
else:
  print("홀수 입니다.")
```

## 입력받은 숫자에 대해 100보다 작거나 100~1000사이거나 1000보다 큰 수자인지 판별하는 프로그램


```python
num = int(num)
if num > 100:
  if num > 1000:
    print("1000이상입니다.")
  else:
    print("100 ~ 1000사이 입니다.")
else:
  print("100보다 작습니다.")
```

## 90점 이상이면 A, 80점이상 B, 70점이상 C, 60점이상 D, 60점이하 F


```python
score = int(input("수학점수를 입력하세요"))
```

```python
if score >=90:
  print("A")
else:
  if score >= 80:
    print("B")
  else:
    if score >= 70:
      print("C")
    else:
      if score >= 60:
        print("D")
      else:
        print("F")
        
  
```
