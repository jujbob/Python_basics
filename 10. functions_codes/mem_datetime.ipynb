## 기념일이 남은 날짜 계산기를 만들어 보자
 - 입력값: 현재시간, 기념일 날짜
 - 출력값: 기념일까지 남은 날짜

```python
from datetime import datetime, timedelta

def getCurrent():
  curDate = datetime.now()
  return curDate
```

```python
def getAfterDate(now, day):
  retDate = now + timedelta(day) 
  return retDate
```

```python
reDate  = getAfterDate(getCurrent(), 100)
```

```python
print("100일후 날짜와 시간", reDate)
```
