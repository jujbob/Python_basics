# 비밀번호 체크하는 프로그램
 - 입력: 비밀번호
 - 조건: 비밀번호가 숫자나 문자로 되어있고 8글자 이상상인지 확인해야함.
 - 출력: 비밀번호가 변경되었는지 아닌지 유무

```python
def checkPassword(pwd):
  if 8 > len(pwd):
    return False
  
  if pwd.isalnum():
    return True
  else:
    return False
```

```python
passwd = ""
```

```python
passwd = input("변경할 비밀번호를 입력하세요: ")
```

```python
if checkPassword(passwd):
  print("비밀번호가 잘 변경 되었어요")
else: 
  print("비밀번호가 변경 규칙에 맞지 않습니다.")
```
