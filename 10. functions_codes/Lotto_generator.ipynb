# 로또 값을 반환하는 함수를 구현하시오
 - 입력: 없음
 - 출력: 1~45의 값중 한개를 반환

```python
import random
```

```python
def lottoFunc():
  num = random.randint(1,45)
  return num
```

```python
lottoList = []
while True:
  lottoNum = lottoFunc()
  if lottoNum in lottoList:
    continue
  else:
    lottoList.append(lottoNum)

  if 6 == len(lottoList):
    break
    
print(lottoList)

extra_num = lottoFunc()
print(extra_num)
```
