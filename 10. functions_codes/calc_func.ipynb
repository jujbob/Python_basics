# 사측연산하는 함수를 구현하시오
 - 입력: 숫자1, 숫자2, 연산유형 (10,20, +)
 - 출력: 연산 결과값

```python
def calc(v1, v2, op):
  result = 0
  if op == '+' :
      result = v1 + v2
  elif op == '-' :
      result = v1 - v2
  elif op == '*' :
      result = v1 * v2
  elif op == '/' :
      result = v1 / v2
  return result
```

```python
cal_val = 0 ##전역변수
var1 = int(input("첫번째 숫자 입력: "))
var2 = int(input("두번째 숫자 입력: "))
operation = input("연산자를 입력: ")

cal_val = calc(var1,var2,operation)
```
