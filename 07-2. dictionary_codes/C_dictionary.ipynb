# 자신이 좋아하는 연예인에 대한 정보를 기입해 오징어게임에 초대하겠다
 - 문제의 정의: 딕셔너리를 활용해서 다음 입력값에 대한 정보를 저장하고 출력하시오
 - 입력: 이름, 구성원수 , 데뷔한곳, 대표곡
 - 출력: 각 value값들

```python
singer = {}
```

```python
singer['이름'] = input("초대할 연예인 이름을 입력하세요: ")
singer['구성원수']= input("초대할 연예인 구성원수을 입력하세요: ")
singer['데뷰']= input("연예인 데뷰을 입력하세요: ")
singer['대표곡']= input("연예인 대표곡을 입력하세요: ")
```

```python
singer
```

```python
singer['데뷰']
```

```python
for k in list(singer.keys()):
  print(k)
```

```python
for v in list(singer.values()):
  print(v)
```
