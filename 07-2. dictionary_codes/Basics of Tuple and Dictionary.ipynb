## 딕셔너리를 언제쓰냐?
 - 김연아: 피겨스케이팅
 - 류현진: 야구
 - 박지성: 축구
 - 손흥민: 축구
 - 김자인: 암벽등반

```python
c_list = ['김연아', '류현진', '박지성', '손흥민']
a_list = ['피겨스케이팅', '야구', '축구', '축구']
```

```python
for idx in c_list:
  if c_list[idx] == '손흥민'
    return idx, a_list[idx]
  else
    return None
```
