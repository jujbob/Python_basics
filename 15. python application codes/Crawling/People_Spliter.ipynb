```python

class Person():
  def __init__(self, department=None, title=None, name=None):
      self.department = department
      self.title = title
      self.name = name

def get_people_list(department, lines):
  p_list = []
  title = ""
  name = ""

  split_index = int(len(lines)/2)
  title_list=lines[:split_index]
  name_list=lines[split_index:]

  for title, name in zip(title_list, name_list):
    people_list.append(Person(department.replace("\n", ""), title.replace("\n", ""), name.replace("\n", "")))

  return p_list


file_name = "texts.txt"

previous_line = ""
department = ""
people_list = []
lines = []

with open(HOME_DIR+file_name, 'r') as file_data:
  for line in file_data:
    if line == "\n" and len(lines) > 0: #build people
      if len(lines)%2 == 1: #data including department
        department = lines[0]
        del lines[0]
        people_list = people_list + get_people_list(department, lines)
        lines = []
      else:  #data not including department
        people_list = people_list + get_people_list(department, lines)
        lines = []

    elif line == "\n":
      previous_line = line
    elif line.startswith("◯") or line.startswith("\ufeff◯"): #만약 ◯出席政府委員와 같은 더미데이터면 삭제
      previous_line = line
    else:
      lines.append(line)
```

```python
for person in people_list:
  print(person.department+"    "+person.title+"  "+person.name)
```
