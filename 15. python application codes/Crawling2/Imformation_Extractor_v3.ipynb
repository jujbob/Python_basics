```python
import pandas as pd

#file_name = "303정무(예산결산기금심사)소위03(11.11.3).txt"
file_name = "315환노(법안심사)소위02(13.4.17)(공청회).txt"
```

```python
def extractor(ex_keyword):
  ex_list = ""
  ex_flag = 0

  with open(HOME_DIR+file_name, 'r', encoding="UTF-8") as file_data:
    for line in file_data:
      
      if line.startswith(ex_keyword) or line.startswith("\ufeff"+ex_keyword):
        ex_flag = 1
      elif ex_flag ==1 and line.startswith("◯") or line.startswith("\ufeff◯"):
        break
      elif ex_flag ==1:
        ex_list += line

  return ex_list
```

```python
attend_s_list = []
all_attend_list = ""
attend_flag = 0
all_attend_flag = 0

with open(HOME_DIR+file_name, 'r', encoding="UTF-8") as file_data:
  for line in file_data:
    
    if line.startswith("◯출석 위원") or line.startswith("\ufeff◯출석 위원"):
      attend_flag = 1
    elif attend_flag ==1 and line.startswith("◯") or line.startswith("\ufeff◯"):
      all_attend_flag = 1
      all_attend_list += line
    elif attend_flag ==1 and all_attend_flag == 1:
      all_attend_list += line
    elif attend_flag ==1:
      attend_s_list += line
```

```python
attend_title_list = [{"id":"조영택", "occupation": "소위원장"}]
```

```python
audi_list = extractor("◯청가 위원")
```

```python
audi_list
```

```python
s_att_list = ''.join(attend_s_list).replace("\n", " ") +" "+ audi_list.replace("\n", "")
print(s_att_list)
```

```python
data = {"파일명": file_name, "출석위원명단": s_att_list, "출석위원직책": attend_title_list, "전체참가자정보":all_attend_list}
df2 = pd.DataFrame(data, columns=['파일명', '부서', '직책', "이름", "전체참가자정보", "출석위원명단", "출석위원직책"])
```
