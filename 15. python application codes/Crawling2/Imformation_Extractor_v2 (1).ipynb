```python
import pandas as pd


file_name = "303정무(예산결산기금심사)소위03(11.11.3).txt"
```

```python

attend_s_list = []
attend_flag = 0

with open(HOME_DIR+file_name, 'r', encoding="UTF-8") as file_data:
  for line in file_data:
    
    
    if line.startswith("◯출석 위원") or line.startswith("\ufeff◯출석 위원"):
      attend_flag = 1
    elif attend_flag ==1 and line.startswith("◯") or line.startswith("\ufeff◯"):
      print(''.join(attend_s_list).replace("\n", " "))
      break
    elif attend_flag ==1:
      #print(line)
      attend_s_list += line


attend_title_list = [{"id":"조영택", "occupation": "소위원장"}]
```

```python

chair_list = ""
chair_flag = 0

with open(HOME_DIR+file_name, 'r', encoding="UTF-8") as file_data:
  for line in file_data:
    
    if line.startswith("◯출석 전문위원") or line.startswith("\ufeff◯◯출석 전문위원"):
      chair_flag = 1
      chair_list += line
    elif chair_flag ==1 and line.startswith("◯") or line.startswith("\ufeff◯"):
      #print(''.join(attend_s_list).replace("\n", " "))
      break
    elif chair_flag ==1:
      #print(line)
      chair_list += line
```

```python

gov_flag = 0
gov_list = ""
with open(HOME_DIR+file_name, 'r', encoding="UTF-8") as file_data:
  for line in file_data:
    
    if line.startswith("◯정부측") or line.startswith("\ufeff◯정부측"):
      gov_flag = 1
      gov_list += line
    elif gov_flag ==1 and line.startswith("   제"):
      #print(''.join(attend_s_list).replace("\n", " "))
      break
    elif gov_flag ==1:
      #print(line)
      gov_list += line
```

```python


data = {"파일명": file_name, "출석위원명단":''.join(attend_s_list).replace("\n", " "), "출석위원직책": attend_title_list, "출석 전문위원":chair_list, "정부측 및 기타 참석자":gov_list}
df2 = pd.DataFrame(data, columns=['파일명', '부서', '직책', "이름", "출석위원명단", "출석위원직책", "출석 전문위원", "정부측 및 기타 참석자"])
```
