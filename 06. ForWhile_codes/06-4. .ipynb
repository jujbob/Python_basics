# 주사위던지기 게임
 - 컴퓨터가 랜덤으로 3개의 주사위를 던져 같은 숫자가 나오면 종료
 - 60번 주사위를 던졌는데 세개의 주사위의 값이 같은숫자가 한번도 나오지 않을경우 "패배하였습니다" 하고 종료하시오

```python
import random

dice1 = 0
dice2 = 0
dice3 = 0

index = 1

while True:
  print(index, "번째 주사위를 던집니다")
  dice1 = random.randint(1,6)
  dice2 = random.randint(1,6)
  dice3 = random.randint(1,6)

  if (dice1 == dice2) and (dice2 == dice3):
    print("3개 주사위는 모두", dice1, "입니다.")
    break  
  else:
    print("각 주사위의 값은", dice1, dice2 ,dice3)
  index = index+1
```
