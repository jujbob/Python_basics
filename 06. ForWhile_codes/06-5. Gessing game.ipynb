# 컴퓨터가 생각한 숫자를 맞추는 게임을 만들어 봅시다.
 - 컴퓨터가 랜덤으로 1~100의 범위내 한개의 숫자를 선택
 - 유져가 특정 숫자를 지정
 - 사용자의 숫자와, 컴퓨터의 숫자가 같을 경우 "맞습니다 축하합니다." 프로그램을 종료하고 틀릴경우 "다시 실행해보세요" 실행 후 다시 입력을 받음
 - **과제1: 사용자가 예측한 숫자가 정답보다 작으면 작다고 알려주고 크면 크다고 알려줍니다.** 
 - **과제2: 사용자가 예측할 수 있는 기회를 10번으로 제한하고, 10번이 넘을경우 "안타깝습니다 패배했습니다"를 출력하고 프로그램 종료** 

```python
import random

comNum = random.randint(1,100)
```

```python
while True:
  userNum = int(input("예측한 숫자를 입력하세요"))
  if comNum==userNum:
    print("맞습니다 축하합니다.")
    break
  else:
    print("다시 시도해 보세요")
```
