## 1.  문제의 정의: 1 ~ 10 까지의 합을 반복문을 활용해 구하시오


```python
sum = 0

for i in range(1, 11, 1):
  sum = sum+i
print(sum)
```

## 2.  문제의 정의: 1000 ~ 2000 까지 홀수의 합을 반복문을 활용해 구하시오


```python
sum = 0

for i in range(1001, 2001, 2):
  sum = sum+i
print(sum)
```

## 3.  문제의 정의: 숫자 두개를 입력으로 받아서, 그 합을 출력하는 프로그램 무한루프 활용한다.
 - 만약 사용자가 첫번째 숫자를 0으로 입력하면 프로그램을 종료한다. 


```python
sum = 0
num1 = 0
num2= 0


#
while True:
  num1 = int(input("첫 번째 숫자를 입력하세요"))
  if num1 == 0:
    break

  num2 = int(input("두 번째 숫자를 입력하세요"))
  sum = num1+num2
  print(sum)
#
print("0을 입력했기 때문에 종료합니다.")

```

## 4.  문제의 정의: 1 ~ 100 까지의 합을 반복문을 활용해 구하시오
 - 단, 4의 배수는 제외 하시오



```python
sum = 0

for i in range(1, 101, 1):
  #print(i)
  if i%4==0:
    continue
  sum = sum+i


print(sum)
```
