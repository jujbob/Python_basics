# 컴퓨터 A, B가 가위바위보 게임을 해서 승률을 반환하는 게임
 - 조건: 컴퓨터는 랜덤으로 가위바위보중 한가지를 선택한다.
 - 조건: 10000번 가위바위보 게임을하여 A,B 각각의 승률과 비긴 횟수를 계산하시오 

```python
import random
```

```python
myList = ["가위", "바위", "보"]
winnerList = []

for myidx in range(10000):
  comA = random.choice(myList)
  comB = random.choice(myList)

  if comA == "가위":
    if comB == "가위":
      print("비겼다")
      winnerList.append("없음")
    elif comB == "바위":
      print("B가 이겼다.")
      winnerList.append("B")
    elif comB == "보" :
      winnerList.append("A")
  elif comA == "바위" :
    if comB == "가위" :
      winnerList.append("A")
    elif comB == "바위" :
      winnerList.append("없음")
    elif comB == "보" :
      winnerList.append("B")
  elif comA == "보" :
    if comB == "가위" :
      winnerList.append("B")
    elif comB == "바위" :
      winnerList.append("A")
    elif comB == "보" :
      winnerList.append("없음")
```
