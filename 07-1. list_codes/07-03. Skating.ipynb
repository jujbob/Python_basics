# 피겨스케이팅 심사 평균점수 결과를 반환하는 코드
  - 조건: 5명의 심사의원이 이 점수를 입력하면 평균점수를 반환함.
  - 조건2: List를 사용한다.

```python

total_score = 0
score_list = []

for i in range(5):
  score_list.append(int(input("점수를 입력하세요")))
```

```python
score_list

for i in range(len(score_list)):
  total_score = total_score + score_list[i]
```

```python
print(total_score/5)
```
