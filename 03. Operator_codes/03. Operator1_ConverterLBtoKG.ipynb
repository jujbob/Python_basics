# Operator 활용 실습 #
 - 개요: 파운드를 키로그램으로, 키로그램을 파운드로 변경하는 프로그램
 - 입력: 키로그램과 파운드를 순차적으로 입력받음
 - 출력: 변환된 키로그램-파운드, 파운드-키로그램 값
 - 조건: 파운드-> 키로그램: 0.453592,   키로그램 --> 파운드: 2.204623

```python
pound = int(input("파운드값을 입력하세요"))
```

```python
out_kg = pound *  0.453592
```

```python
print("pound값: ", pound, "   변환된 kg값: ", out_kg)
```
