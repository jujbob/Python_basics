# Operator 활용 실습 #
 - 개요: 내가 수강한 과목의 시수와 받은 최종 학점을 토대로 학점 평균을 구하는 프로그램 
 - 입력: 내가 수강한 과목에서 받은 학점
 - 출력: 학점평균
 - 조건: 평균 = 획득한 학점/총 학점

```python
python = 3
mobile = 2
ai = 1
```

```python
"""
A = 4.5
A0 = 4.0
B = 3.5
"""
```

```python
userName = input("너의 이름은?: ")
python_score= int(input("파이썬 과목에서 획득한 점수를 입력 하세요"))
mobile_score= int(input("모바일 과목에서 획득한 점수를 입력 하세요"))
ai_score= int(input("AI 과목에서 획득한 점수를 입력 하세요"))
```

```python
weight_sum = (python * python_score) + (mobile * mobile_score) + (ai * ai_score)
```

```python
total_degree = python+mobile+ai
```

```python
total_degree
```

```python
average = weight_sum/total_degree
```

```python
print("평균학점은: ", average)
```
